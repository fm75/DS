```python
import requests
import pandas as pd

def get_daily_csv(url: str) -> pd.DataFrame:
    header = {
      "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
      "X-Requested-With": "XMLHttpRequest"
    }
    r = requests.get(url, headers=header).text
    return pd.read_html(r, header=[0], index_col=1)[0]
```

```python
def covid19_repo() -> str:
    return 'https://github.com/CSSEGISandData/COVID-19'


def daily_csv(date: str) -> str:
    return '/'.join([covid19_repo(), f'tree/master/csse_covid_19_data/csse_covid_19_daily_reports/{date}.csv'])

```

```python
df = get_daily_csv(daily_csv('03-29-2020'))
```
