```python
import requests
```

```python
url = 'https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data/csse_covid_19_daily_reports/03-27-2020.csv'
```

```python
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

r = requests.get(url, headers=header)
```

```python
df = pd.read_html(r.text, header=[0], index_col=0)[0].iloc[:-1]
```

```python
df[df.Country_Region=='US'].sort_values(by='Province_State')
```

```python
df[df.Province_State=='New Jersey'].sort_values(by='Confirmed')
```

```python
df[df.Province_State=='New Hampshire'].sort_values(by='Confirmed')
```
